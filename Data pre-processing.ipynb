```python
import pandas as pd

df = pd.read_csv('Dataset.csv')
```

```python

# Create a new DataFrame with only 'Tarih/Bölge' column
new_df = pd.DataFrame(columns=['Başakşehir','Esenyurt','Kağıthane','Kandilli','Mecidiyeköy','Silivri']) 

# Iterate over each row and column
for index in range(len(df)):
    for col_idx, col in enumerate(df.columns[1:], start=1):  # Start from index 1 to skip 'Tarih/Bölge'
        if df.at[index, col] == '-':
            print("Found '-' value at row {}, column {}".format(index, col))
            # Replace '-' with the corresponding value from 'Tarih/Bölge'
            new_df.at[index, col] = df.at[index, 'Tarih/Bölge']
            
new_df.to_csv('updated_dataset.csv', index=False)
```
